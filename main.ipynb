```python
def commonpokemon(df):
    df2 = df["name"].value_counts()

    df3 = df2.reset_index(name = "count")
    df3.rename(columns={"index":"name"}, inplace= True)
    fig = px.bar(df3.head(10),y = "name",x = "count", orientation='h', title = "Most Common Pokemon")
    return fig
def commontype(df):
    df2 = df["type1"].value_counts()

    df3 = df2.reset_index(name = "count")

    dftype2 = df["type2"].value_counts()

    dftype3 = dftype2.reset_index(name = "count")

    df3 = df3.merge(dftype3, on = "index")

    df3.rename(columns={"index":"name","count_x": "first_type","count_y": "second_type"}, inplace= True)

    fig = px.bar(df3,y = "name",x = ["first_type","second_type"], orientation='h',title = "Most Common Types")
    return fig
def map(df):
    fig = px.scatter_geo(df,
                    lat="lat",
                    lon="long",
                    hover_name="name",
                    color = "name",
                    title = "Locations of Pokemon")
    fig.update_layout(showlegend=False)
    return fig
```
